# Exercise 9.5: Singular values

```python
import numpy as np

def exercise():
    for n in range(3,210,3):
        matrix_C = np.zeros((n,n))
        for i in range(1,10,2):
            p = i/10
            for k in range(n):
                for l in range(n):
                    if np.random.random() > 1-p:
                        matrix_C[k][l] = 1
                    else:
                        matrix_C[k][l] = 0
            n1,n2,n3 = np.linalg.svd(matrix_C)
            max_singular_value = max(n2)
            print("The Largest Singular Value = ",max_singular_value,"p = ",p,"n = ",n)

if __name__ == "__main__":
    exercise()
```

# 10. Scipy

# Exercise 10.1: Least squares

```python
import numpy as np
import scipy.optimize as opt

m, n = 100, 50
# Generate matrix A ∈ Rm × n with m > n
A = np.random.rand(m, n)  

# generate some vector b ∈ Rm.
b = np.random.random(m)

# find x = argminx k Ax − bk 2.
res = opt.lsq_linear(A, b)
x = res.x

# Print the norm of the residual.
residual = b - np.dot(A, x)
norm = np.linalg.norm(residual)

print(norm)
```

# Exercise 10.2: Optimization

```python
import numpy as np
import scipy.optimize as opt


def f(x):  
    res = np.power(np.sin(x-2), 2) * np.exp(-1*(x**2))

    return -1 * res
    
res = opt.minimize_scalar(f)
x = res.x
print(-1*f(x))
```

# Exercise 10.3: Pairwise distances

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.spatial.distance as dis

n, m = 50, 100
# Let X be a matrix with n rows and m columns
X = np.random.rand(n, m)

# The first element is dist(X[0], X[1])
# the second is dist(X[0], X[2]) 
# the third is dist(X[0], X[3])
de = dis.pdist(X)

# As an example application, consider n cities, and we are given their coordinates in two columns
x = []
y = []
for a in range(0, n//2):
	x = x + [a for _ in range(1, n)]
	y = y + [b for b in range(1, n)]

x = np.array(x) # set x lable
y = np.array(y) # set y lable


fig = plt.figure()
ax = Axes3D(fig)

# creat a 3-D table
plt.title("distance between two cities")  
ax.set_xlabel('first city') 
ax.set_ylabel('second city')
ax.set_zlabel('distance')

ax.scatter(x, y, de, c='r')
plt.show()
```

# 11. Matplotlib

# Exercise 11.1: Plotting a function

```python
import matplotlib.pyplot as plt
import math
import numpy as np


def fun(x):
    return math.sin(x-2) ** 2 * math.exp(-x ** 2)


x = np.linspace(0, 2, num=1000)
plt.style.use('seaborn')
fun = np.vectorize(fun)
plt.plot(x, fun(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x)=sin^2(x)*(x-2)*e^{-x^2}$')
plt.show()
```

# Exercise 11.2: Data

```python
import matplotlib.pyplot as plt
import math
import numpy as np

X = np.random.normal(0, 1, [20, 10])
b = np.random.normal(0, 1, 10)
z = np.random.normal(0, 1, 20)
y = np.matmul(X, b.reshape(-1, 1)) + z.reshape(-1, 1)

inverse = np.linalg.inv(np.matmul(np.transpose(X), X))
w = np.matmul(inverse, np.matmul(np.transpose(X), y))

x = np.linspace(1, 10, 10, endpoint=True)
plt.scatter(x, b, marker='x', color='r',label='True coefficients')
plt.scatter(x, w, marker='o', color='blue',label='Estimated cofficients')
plt.legend(loc=1, ncol=1, prop={'size': 7})
plt.plot(x, [0.0 for i in range(10)], color='gray', alpha=0.5)
plt.xticks(x)
plt.margins(x=0)
plt.show()
```

# Exercise 11.3: Histogram and density estimation

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

n_basesample = 1000

alpha = 0.6 
mlow, mhigh = (-3.0, 3.0) 
xn = np.concatenate([mlow + np.random.randn(int(alpha * n_basesample)),
                     mhigh + np.random.randn(int((1 - alpha) * n_basesample))])

# Generate core
gkde = stats.gaussian_kde(xn)

ind = np.linspace(-7, 7, 101)
kdepdf = gkde.evaluate(ind)

plt.style.use('seaborn')

# plot histgram of sample
plt.hist(xn, bins=25, density=1)

# plot estimated density
plt.plot(ind, kdepdf, label='kde', color="g")

# plot data generating density
plt.plot(ind, alpha * stats.norm.pdf(ind, loc=mlow) +
         (1 - alpha) * stats.norm.pdf(ind, loc=mhigh),
         color="r", label='normal mix')

plt.title('Kernel Density Estimation')
plt.legend()
plt.xticks()
plt.show()
```
